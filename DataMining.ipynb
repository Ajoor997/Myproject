```python
import pandas as np
import numpy as np
import bs4
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
import requests

link = 'https://www.metacritic.com/browse/games/release-date/available/pc/metascore?page='
hdr = {'User-Agent':'Mozilla/5.0'}
Pages = 28

for i in range(Pages-1):
    html = requests.get(link+str (i), headers=hdr).content
    PageSoup = soup(html,"html.parser")

    Container1 = PageSoup.findAll("li", {"class":"product game_product"})
    First = PageSoup.findAll("li", {"class":"product game_product first_product"})
    Second = PageSoup.findAll("li", {"class":"product game_product last_product"})
    Container1 = First + Container1 + Second

    Containers = Container1
    Container = Container1[0]

    y = len (Containers)
    for x in range(y):
        Containers[x].find("span", {"class":"label"}).decompose()
    
    filename = "VideoGames.csv"
    f = open(filename, 'a')

    for Container in Containers:
        title = Container.div.div.text.strip()
        title = title.replace(",","")
        
        metacritic_score = Container.findAll("div", {"class":"metascore_w small game positive"})
        user_score = Container.findAll("li", {"class":"stat product_avguserscore"})
        
        if (len(metacritic_score) ==0):
             metacritic_score = Container.findAll("div", {"class":"metascore_w small game mixed"})
        
        if (len(metacritic_score) ==0):
             metacritic_score = Container.findAll("div", {"class":"metascore_w small game negative"})
                
        if (len(metacritic_score) ==0):
             metacritic_score = Container.findAll("div", {"class":"metascore_w medium game negative"})
       
        metacritic_score =  metacritic_score[0].text
        user_score = user_score[0].span.text
            
        f.write(title + "," +  metacritic_score + "," + user_score + "\n")  
    f.close()
    print (i, "Imported")

print ("Completed")
```
